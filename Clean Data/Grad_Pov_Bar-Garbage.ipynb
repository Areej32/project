```python
# dependencies
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
# import matplotlib.pyplot as plt
```

```python
# load excel file
# File to Load 
file_to_load = "County Grad Rate with Poverty Levels.csv"


# Read Purchasing File and store into Pandas data frame
county_grad_pov = pd.read_csv(file_to_load)
```

```python
df_county_grad_pov = pd.DataFrame(county_grad_pov)
df_county_grad_pov
```

```python
county = df_county_grad_pov["County Name"]
grad_rate = df_county_grad_pov["Graduation Rate"]
x_axis = np.arange(len(grad_rate))
```

```python
# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, county, rotation= 90)
plt.bar(x_axis, grad_rate, color="b", align="center")
plt.title("County Graduation Rates")
plt.xlabel("Counties")
plt.ylabel("Graduation Rates")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Graduation_Rates.png", bbox_inches = "tight")
plt.show()
```

```python
county = df_county_grad_pov["County Name"]
pov_rate = df_county_grad_pov["Poverty Rate"]
x_axis = np.arange(len(pov_rate))
# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, county, rotation= 90)
plt.bar(x_axis, pov_rate, color="r", align="center")
plt.title("County Poverty Rates")
plt.xlabel("Counties")
plt.ylabel("Poverty Rates")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Poverty_Rates.png", bbox_inches = "tight")
plt.show()
```
