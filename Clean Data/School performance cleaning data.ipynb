```python
# dependencies
import pandas as pd
import os
import matplotlib
```

```python
# load excel file
# File to Load 
file_to_load = "2017 Grad Rates_4Year.csv"


# Read Purchasing File and store into Pandas data frame
Schools_performance_data = pd.read_csv(file_to_load)
# Schools_performance_data.head()
```

```python
# remove all NaN values
Schools_performance_data.dropna(inplace=True)



# replace blank values with zeros
Schools_performance_data.replace('-', 0, inplace = True) 

print(Schools_performance_data.count())

```

```python
Schools_performance_df=pd.DataFrame(Schools_performance_data)
Schools_performance_df.rename(columns={"COUNTY_CODE": "County Code"},inplace=True)
Schools_performance_df.rename(columns={"COUNTY_NAME": "County Name"},inplace=True)
Schools_performance_df.rename(columns={"DISTRICT_CODE": "District Code"},inplace=True)
Schools_performance_df.rename(columns={"DISTRICT_NAME": "District Name"},inplace=True)
Schools_performance_df.rename(columns={"SCHOOL_CODE": "School Code"},inplace=True)
Schools_performance_df.rename(columns={"SCHOOL_NAME": "School Name"},inplace=True)
Schools_performance_df.rename(columns={"SUBGROUP": "Subgroup"},inplace=True)
Schools_performance_df.rename(columns={"FOUR_YR_GRAD_RATE": "Four Years Graduation Rate"},inplace=True)
Schools_performance_df.rename(columns={"FOUR_YR_ADJ_COHORT_COUNT": "Four Years Adjusted Cohort count "},inplace=True)
Schools_performance_df.rename(columns={"GRADUATED_COUNT": "Graduated Count"},inplace=True)





Schools_performance_df.head()
```
