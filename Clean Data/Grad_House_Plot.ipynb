```python
# dependencies
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File to Load 
file_to_load = "County Grad Rate with Household Size.csv"


# Read Purchasing File and store into Pandas data frame
county_grad_pov = pd.read_csv(file_to_load)
```

```python
df_county_grad_house = pd.DataFrame(county_grad_pov)
df_county_grad_house
```

```python
county = df_county_grad_house["County Name"]
house_size = df_county_grad_house["HouseHold Size"]
x_axis = np.arange(len(house_size))
# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, county, rotation= 90)
plt.bar(x_axis, house_size, color="r", align="center")
plt.title("County Household Size")
plt.xlabel("Counties")
plt.ylabel("HOusehold Size")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Household_Size.png", bbox_inches = "tight")
plt.show()
```

```python
grad_rate = df_county_grad_house["Graduation Rate"]
house_size = df_county_grad_house["HouseHold Size"]

plt.scatter(house_size, grad_rate)

plt.grid()
plt.title("NJ High School Graduation Rates and Household Size by County")
plt.ylabel("Graduation Rates")
plt.xlabel("Household Size")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Grad_House_Rates.png", bbox_inches = "tight")
plt.show()
```

```python
fig, ax1 = plt.subplots()

tick_locations = [value for value in x_axis]

plt.xticks(tick_locations, county, rotation= 90)

grad_rate = df_county_grad_house["Graduation Rate"]
county = df_county_grad_house["County Name"]
pov_rate = df_county_grad_house["HouseHold Size"]
t = np.arange(len(county))
ax1.plot(t, pov_rate, 'b-')
ax1.set_xlabel('county')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('HouseHold Size', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()

ax2.plot(t,grad_rate, 'r*')
ax2.set_ylabel('Graduation Rate', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
plt.savefig("Images/grad vs household.jpg", bbox_inches = "tight")
```
