```python
# dependencies
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# File to Load 
file_to_load1 = "County Grad Rate with Poverty Levels.csv"


# Read Purchasing File and store into Pandas data frame
county_grad_pov = pd.read_csv(file_to_load1)
```

```python
df_county_grad_pov = pd.DataFrame(county_grad_pov)
df_county_grad_pov
```

```python
#Setup Axis
county = df_county_grad_pov["County Name"]
grad_rate = df_county_grad_pov["Graduation Rate"]
x_axis = np.arange(len(grad_rate))

# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, county, rotation= 90)
plt.bar(x_axis, grad_rate, color="b", align="center")
plt.title("County Graduation Rates")
plt.xlabel("Counties")
plt.ylabel("Graduation Rates")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Graduation_Rates.png", bbox_inches = "tight")
plt.show()
```

```python
county = df_county_grad_pov["County Name"]
pov_rate = df_county_grad_pov["Poverty Rate"]
x_axis = np.arange(len(pov_rate))
# Create a bar chart based upon the above data
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, county, rotation= 90)
plt.bar(x_axis, pov_rate, color="r", align="center")
plt.title("County Poverty Rates")
plt.xlabel("Counties")
plt.ylabel("Poverty Rates")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Poverty_Rates.png", bbox_inches = "tight")
plt.show()
```

```python
grad_rate = df_county_grad_pov["Graduation Rate"]
pov_rate = df_county_grad_pov["Poverty Rate"]

plt.scatter(pov_rate, grad_rate)

plt.grid()
plt.title("NJ High School Graduation Rates and Poverty Rates by County")
plt.ylabel("Graduation Rates")
plt.xlabel("Poverty Rate")

# Save an image of the chart and print it to the screen
plt.savefig("Images/County_Grad_Poverty_Rates.png", bbox_inches = "tight")
plt.show()
```

```python
fig, ax1 = plt.subplots()

tick_locations = [value for value in x_axis]

plt.xticks(tick_locations, county, rotation= 90)

grad_rate = df_county_grad_pov["Graduation Rate"]
county = df_county_grad_pov["County Name"]
pov_rate = df_county_grad_pov["Poverty Rate"]
t = np.arange(len(county))
ax1.plot(t, pov_rate, 'b-')
ax1.set_xlabel('county')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('poverty rate', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()

ax2.plot(t,grad_rate, 'r*')
ax2.set_ylabel('Graduation Rate', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
plt.savefig("PyberBubblePlot.jpg", bbox_inches = "tight")
```

```python
fig = plt.figure(figsize=(8,6)) 
x = df_county_grad_pov["Poverty Rate"] 
y = df_county_grad_pov["Graduation Rate"] 
plt.scatter(x, y, color="g", marker="o", alpha=0.9) 

#Calculate and add R2 value
mask = ~np.isnan(x) & ~np.isnan(y)


#Add regression line
sns.regplot(df_county_grad_pov["Poverty Rate"], 
              df_county_grad_pov["Graduation Rate"], color='r',label = "Poverty " )

# Incorporate the other graph properties
plt.title("Correlation between Graduation rates and Poverty rates")
plt.ylabel("Graduation rate")
plt.xlabel("Poverty rate") 

plt.legend(loc='best')
plt.grid(True)
sns.set_style('whitegrid')
plt.text(21, 0.925, "Note:\nWe notice an inverse corellation ship between Graduation rates and Povetry rates.")


plt.show()
```
