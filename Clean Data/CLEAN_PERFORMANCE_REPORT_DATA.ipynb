```python
# dependencies
import pandas as pd
import os
import matplotlib
```

```python
# File to Load 
file_to_load = "2017_PerformanceReports.csv"


# Read Purchasing File and store into Pandas data frame
performance_report = pd.read_csv(file_to_load)
performance_report.head(450)
```

```python
# remove all NaN values
performance_report.dropna(inplace=True)



# replace blank values with zeros
performance_report.replace('-', 0, inplace = True) 

print(performance_report.count())
```

```python
performance_report_df=pd.DataFrame(performance_report)
performance_report_df.rename(columns={"COUNTY_CODE": "County Code"},inplace=True)
performance_report_df.rename(columns={"DISTRICT_CODE": "District Code"},inplace=True)
performance_report_df.rename(columns={"SCHOOL_CODE": "School Code"},inplace=True)
performance_report_df
```

```python
performance_report_df.to_csv("CLEAN_PERFORMANCE_REPORT_DATA.csv")
```
